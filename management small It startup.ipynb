```python
class Task:
    def __init__(self, name, assigned_to, deadline):
        self.name = name
        self.assigned_to = assigned_to
        self.deadline = deadline
        self.status = "Pending"

    def complete_task(self):
        self.status = "Completed"

    def __str__(self):
        return f"Task: {self.name}, Assigned to: {self.assigned_to}, Deadline: {self.deadline}, Status: {self.status}"


class Project:
    def __init__(self, project_name):
        self.project_name = project_name
        self.tasks = []

    def add_task(self, task):
        self.tasks.append(task)

    def show_tasks(self):
        print(f"\nProject: {self.project_name}")
        for task in self.tasks:
            print(task)

    def project_status(self):
        completed = sum(1 for task in self.tasks if task.status == "Completed")
        total = len(self.tasks)
        print(f"\nProject Progress: {completed}/{total} tasks completed")


# ---- Example Usage in Small IT Startup ---- #

# Create Project
startup_project = Project("Website Development")

# Add Tasks
task1 = Task("Design Homepage", "Alice", "2026-03-01")
task2 = Task("Develop Backend API", "Bob", "2026-03-10")
task3 = Task("Testing & Debugging", "Charlie", "2026-03-15")

startup_project.add_task(task1)
startup_project.add_task(task2)
startup_project.add_task(task3)

# Show Tasks
startup_project.show_tasks()

# Complete a Task
task1.complete_task()

# Show Updated Status
startup_project.project_status()
```

```python
# Project Management System for Small IT Startup

class Employee:
    def __init__(self, emp_id, name, role):
        self.emp_id = emp_id
        self.name = name
        self.role = role

    def __str__(self):
        return f"{self.emp_id} - {self.name} ({self.role})"


class Task:
    def __init__(self, task_id, task_name, deadline, assigned_to):
        self.task_id = task_id
        self.task_name = task_name
        self.deadline = deadline
        self.assigned_to = assigned_to
        self.status = "Pending"

    def update_status(self, status):
        self.status = status

    def __str__(self):
        return f"{self.task_id} | {self.task_name} | {self.deadline} | {self.assigned_to.name} | {self.status}"


class Project:
    def __init__(self, project_name):
        self.project_name = project_name
        self.tasks = []

    def add_task(self, task):
        self.tasks.append(task)

    def show_tasks(self):
        print(f"\nProject: {self.project_name}")
        for task in self.tasks:
            print(task)

    def project_progress(self):
        completed = sum(1 for task in self.tasks if task.status == "Completed")
        total = len(self.tasks)
        print(f"Progress: {completed}/{total} tasks completed")


# ---- Main Program ---- #

employees = []
projects = []

def add_employee():
    emp_id = input("Enter Employee ID: ")
    name = input("Enter Employee Name: ")
    role = input("Enter Role: ")
    emp = Employee(emp_id, name, role)
    employees.append(emp)
    print("Employee added successfully!\n")


def create_project():
    name = input("Enter Project Name: ")
    project = Project(name)
    projects.append(project)
    print("Project created successfully!\n")


def add_task_to_project():
    if not projects:
        print("No projects available!\n")
        return

    for i, project in enumerate(projects):
        print(i, "-", project.project_name)

    p_index = int(input("Select project number: "))
    project = projects[p_index]

    task_id = input("Enter Task ID: ")
    task_name = input("Enter Task Name: ")
    deadline = input("Enter Deadline (YYYY-MM-DD): ")

    if not employees:
        print("No employees available!\n")
        return

    for i, emp in enumerate(employees):
        print(i, "-", emp)

    e_index = int(input("Select employee number: "))
    assigned_emp = employees[e_index]

    task = Task(task_id, task_name, deadline, assigned_emp)
    project.add_task(task)
    print("Task added successfully!\n")


def update_task_status():
    for project in projects:
        for task in project.tasks:
            print(task)
            new_status = input("Enter new status (Pending/In Progress/Completed): ")
            task.update_status(new_status)
    print("Task status updated!\n")


def show_all_projects():
    for project in projects:
        project.show_tasks()
        project.project_progress()


# ---- Menu ---- #

while True:
    print("------ Small IT Startup Project Management ------")
    print("1. Add Employee")
    print("2. Create Project")
    print("3. Add Task to Project")
    print("4. Update Task Status")
    print("5. Show All Projects")
    print("6. Exit")

    choice = input("Enter choice: ")

    if choice == "1":
        add_employee()
    elif choice == "2":
        create_project()
    elif choice == "3":
        add_task_to_project()
    elif choice == "4":
        update_task_status()
    elif choice == "5":
        show_all_projects()
    elif choice == "6":
        print("Exiting system...")
        break
    else:
        print("Invalid choice!\n")
```
